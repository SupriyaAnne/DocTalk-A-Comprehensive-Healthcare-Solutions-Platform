```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix ,r2_score, mean_squared_error
import numpy as np
from matplotlib import pyplot as plt

iris = datasets.load_iris()
features = iris.data[:100, :2]
target = iris.target[:100]

features_train, features_test, target_train, target_test = train_test_split(
features, target, test_size=0.2, random_state=42
)

scaler = StandardScaler()
features_train_standardized = scaler.fit_transform(features_train)
features_test_standardized = scaler.transform(features_test)

svc = LinearSVC(C=1.0)
model = svc.fit(features_train_standardized, target_train)


color = ["black" if c == 0 else "red" for c in target_train]
plt.scatter(features_train_standardized[:, 0], features_train_standardized[:, 1], c=color)
w = svc.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-2.5, 2.5)
yy = a * xx - (svc.intercept_[0]) / w[1]
plt.plot(xx, yy, label="Decision Boundary")
plt.title("Linear SVM Decision Boundary ")
plt.xlabel("Feature 1 ")
plt.ylabel("Feature 2 ")
plt.legend()
plt.show()


predictions = svc.predict(features_test_standardized)

accuracy = accuracy_score(target_test, predictions)
conf_matrix = confusion_matrix(target_test, predictions)

print("Accuracy: {:.2f}".format(accuracy))
print("R2 score",r2_score(target_test, predictions))
print("MSE", mean_squared_error(target_test, predictions))
print("\nConfusion Matrix:\n", conf_matrix)


new_observation = [[-2, 3]]

svc.predict(scaler.transform(new_observation))
```
